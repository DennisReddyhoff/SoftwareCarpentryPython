# Lesson 6: Analysing Data from Multiple Files

```python
import glob
import numpy as np
import matplotlib.pyplot as mpl
```

```python
print(glob.glob('../data/inflammation*.csv'))
```

We can use glob to pull in the first 3 files and plot them. We use the sorted function to sort the list

```python
filenames = sorted(glob.glob('../data./inflammation*.csv'))
filenames = filenames[0:3]
for filename in filenames:
    print(filename)
    
    data = np.loadtxt(fname=filename, delimiter=',')
    
    fig = mpl.figure(figsize=(10, 3))
    
    axes1 = fig.add_subplot(1, 3, 1)
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1 ,3, 3)
    
    axes1.set_ylabel('average')
    axes1.plot(np.mean(data, axis=0))
    
    axes2.set_ylabel('max')
    axes2.plot(np.max(data, axis=0))
    
    axes3.set_ylabel('min')
    axes3.plot(np.min(data, axis=0))
    
    fig.tight_layout()
    mpl.show()
```

We can plot the difference of two datasets

```python
fig2 = mpl.figure(figsize=(10.0, 3.0))

data0 = np.loadtxt(fname=filenames[0], delimiter=',')
data1 = np.loadtxt(fname=filenames[1], delimiter=',')

mpl.ylabel('Difference in average')
mpl.plot(np.mean(data0, axis=0) - np.mean(data1, axis=0))

fig2.tight_layout()
mpl.show()
```

We can generate summary statistics for the datasets

```python
composite_data = np.zeros((60,40))

for filename in filenames:
    data = np.loadtxt(fname=filename, delimiter=',')
    composite_data = composite_data + data
    
composite_data = composite_data / len(filenames)

fig3 = mpl.figure(figsize=(10, 3))
    
axes1 = fig3.add_subplot(1, 3, 1)
axes2 = fig3.add_subplot(1, 3, 2)
axes3 = fig3.add_subplot(1 ,3, 3)
    
axes1.set_ylabel('average')
axes1.plot(np.mean(composite_data, axis=0))
    
axes2.set_ylabel('max')
axes2.plot(np.max(composite_data, axis=0))
    
axes3.set_ylabel('min')
axes3.plot(np.min(composite_data, axis=0))
    
fig3.tight_layout()
mpl.show()
```
