````python
import numpy as np
import matplotlib.pyplot as mpl
import glob
````

````python
def visualise_data(filename):
    
    data = np.loadtxt(fname=filename, delimiter=',')
    
    fig = mpl.figure(figsize=(10.0, 3.0))
    
    axes1 = fig.add_subplot(1, 3, 1)
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)
    
    axes1.set_ylabel('average')
    axes1.plot(np.mean(data, axis=0))
    
    axes2.set_ylabel('max')
    axes2.plot(np.max(data, axis=0))
    
    axes3.set_ylabel('min')
    axes3.plot(np.min(data, axis=0))
    
    fig.tight_layout()
    mpl.show()
````

````python
def detect_problems(filename):
    
    data = np.loadtxt(fname=filename, delimiter=',')

    if np.max(data, axis=0)[0] == 0 and np.max(data,axis=0)[20] == 20:
        print('Suspicious looking maxima')
    elif np.sum(np.min(data, axis=0)) == 0:
        print('Minima add to zero!')
    else:
        print('Looks ok')
    
````

````python
filenames = sorted(glob.glob("../data/inflammation*.csv"))

for filename in filenames[:3]:
    print(filename)
    visualise_data(filename)
    detect_problems(filename)
````

Once we start putting things in functions so that we can re-use them, we need to start testing that those functions are working correctly. To see how to do this, let’s write a function to offset a dataset so that it’s mean value shifts to a user-defined value:

````python
def offset_mean(data, target_mean_value):
    return (data - np.mean(data)) + target_mean_value
````

````python
z = np.zeros((2,2))
print(offset_mean(z, 3))
````

````python
data = np.loadtxt(fname="../data/inflammation-01.csv", delimiter=",")
print(offset_mean(data, 0))
````

````python
print("Original min, mean and max:", np.min(data), np.mean(data), np.max(data))
offset_data = offset_mean(data, 0)
print("Offset min, mean and max: ", np.min(offset_data), np.mean(offset_data),
      np.max(offset_data))
````

````python
print("Difference in standard deviation before and after:",
      np.std(data) - np.std(offset_data))
````

````python
def offset_mean(data, target_mean_value=0.0):
    '''Return a new array containing original data
    with it's mean offset to match desired value (0 by default)
    
    Parameters
    ---------- 
    data : numpy.ndarray
        Numpy array of data to be offset
    target_mean_value : float, optional
        New mean value of `data` (default=0)
        
    Returns
    -------
    numpy.ndarray
        (data - np.mean(data)) + target_mean_value
    
    Examples
    --------
    >>>offset_mean([1, 2, 3])
    array([-1, 0 , 1])
    '''
    
    return (data - np.mean(data)) + target_mean_value
    
    
````

````python
test_data = np.zeros((2,2))
print(offset_mean(test_data, 3))
````

````python
more_data = 5 + np.zeros((2,2))
print('data before offset: \n', more_data)
print('offset data: \n', offset_mean(more_data))
````

Combining Strings
“Adding” two strings produces their concatenation: 'a' + 'b' is 'ab'. Write a function called fence that takes two parameters called original and wrapper and returns a new string that has the wrapper character at the beginning and end of the original. A call to your function should look like this:

```>>>print(fence('name', '*'))
*name*```

````python
def fence(original, wrapper):
    ''' Wraps a string with a user input character
    
    Parameters
    ---------- 
    original : str
        String to be wrapperd
    wrapper : str
        Character to wrap string with
        
    Returns
    ------- 
    new_string : str 
        Wrapped string
    '''
    
    new_string = wrapper + original + wrapper
    
    return new_string
````

````python
fence('name', '*')
````

````python
def outer(user_string):
    return user_string[0] + user_string[-1]
````

````python
def rescale(user_array):
    array_low = np.min(user_array)
    array_high = np.max(user_array)
    returned_array = (user_array - array_low) / (array_high - array_low)
    return returned_array
````

````python
x = np.array([[1, 2, 3],
              [7, 4, 2],
              [8, 6, 3]])

rescale(x)

y = np.reshape(np.arange(9), (3,3))

rescale(y)
````

````python
def rescale(user_array, low_lim=0.0, high_lim=1.0):
    array_low = np.min(user_array)
    array_high = np.max(user_array)
    rescaled_array = (user_array - array_low) / (array_high - array_low)
    returned_array = rescaled_array * (high_lim - low_lim) + low_lim
    return returned_array
````

````python
rescale(x)
rescale(x, 1, 2)
````
