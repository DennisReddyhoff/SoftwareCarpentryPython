```python
# Use magic command to enable inline plotting
%matplotlib inline
import matplotlib.pyplot as mpl
```

```python
time = [0, 1, 2, 3]
pos = [0, 100, 200, 300]

mpl.plot(time, pos)
mpl.xlabel('Time (hr)')
mpl.ylabel('Position (km)')
```

We can plot from a pandas dataframe

```python
import pandas as pd
import seaborn
```

```python
data = pd.read_csv("../../data/gapminder_gdp_oceania.csv",
                   index_col="country")

# Strip all but year from column names
years = data.columns.str.strip('gdpPercap_')
# Convert to integers
data.columns = years.astype(int)

data.loc['Australia'].plot()
```

```python
data.T.plot()
mpl.ylabel('GDP per capita')
```

```python
mpl.style.use('seaborn-dark')
data.T.plot(kind = 'bar')
mpl.ylabel('GDP per capita')
```

We can plot directly with mpl.plot(x, y)

```python
year = data.columns
gdp_australia = data.loc["Australia"]
mpl.plot(year, gdp_australia, "g--")
```

```python
gdp_nz = data.loc["New Zealand"]

mpl.plot(year, gdp_australia, "g--")
mpl.plot(year, gdp_nz, "b--")
```

```python
# Add a legend after labelling datasets
mpl.plot(year, gdp_australia, label="Australia")
mpl.plot(year, gdp_nz, label="New Zealand")
mpl.legend()
```

```python
mpl.scatter(gdp_australia, gdp_nz)
```

```python
data.T.plot.scatter(x = "Australia", y = "New Zealand")
```

```python
data_europe = pd.read_csv("../../data/gapminder_gdp_europe.csv", 
                          index_col = "country")
data_europe.min().plot(label = "min")
data_europe.max().plot(label = "max")
mpl.legend(loc = "best")
mpl.xticks(rotation=90)
```

```python
data_asia = pd.read_csv("../../data/gapminder_gdp_asia.csv", 
                        index_col = "country")
data_asia.describe().T.plot(kind = "scatter",
                            x = "min", y = "max")
```

```python
data_asia.max().plot()
```

```python
print(data_asia.idxmax())
print(data_asia.idxmin())
```

```python
mpl.style.use("seaborn")
data_all = pd.read_csv("../../data/gapminder_all.csv", index_col = "country")
data_all.plot(kind = "scatter", 
              x = "gdpPercap_2007",
              y = "lifeExp_2007", 
              s = data_all["pop_2007"]/1e6)
mpl.gcf().savefig("../../data/figure1.svg")
```
